```python
camera = cv.VideoCapture(0)
print("BEFORE")
while (camera.isOpened()):
  print("OPEN")
  ret, frame = camera.read()
  cv.imshow('webCam',frame)
  if (cv.waitKey(1) == ord('s')):
    break
print("AFTER")

camera.release()
cv.destroyAllWindows()
```
